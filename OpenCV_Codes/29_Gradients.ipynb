```python
import cv2
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('Computer-Vision-with-Python/DATA/sudoku.jpg')
```

```python
display_img(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize = 5)
```

```python
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize = 5)
display_img(sobely)
```
